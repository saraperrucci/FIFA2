```python
def clean_df(df):
    df.columns = df.columns.str.replace(' ', '_').str.lower()
    df = df.drop_duplicates()
    columns_with_nan = df.columns[df.isna().any()]
    df = df[df['value'] != '€0']
    players_with_value_0 = df[df['value'] == '€0']
    columns_with_nan = df.columns[df.isna().any()]
    composure_nan = df[df['composure'].isna()]
    def fill_missing_numerical(df):
        numerical_columns = df.select_dtypes(include=['number']).columns
        for col in numerical_columns:
            median_value = df[col].median()
            df[col].fillna(median_value, inplace=True)
        fill_missing_numerical(df)
    columns_with_nan = df.columns[df.isna().any()]
    df = df.drop(columns=['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf',
       'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb',
       'cb', 'rcb', 'rb', 'gk'])
    df = df.rename(columns={'bp':'best_position', 'team_&_contract' :'team_and_contract', 'w/f' : 'weak_foot', 'sm' : 'skill_moves', 'a/w' : 'attacking_work_rate', 'd/w' : 'defensive_work_rate', 'ir' : 'international_reputation', 'pac' : 'pace', 'sho' : 'shooting', 'pas' : 'passing', 'dri' : 'dribbing', 'def' : 'defending', 'phy' : 'physicality', 'ova' : 'overall_rating', 'fk_accuracy' : 'free_kick_accuracy', 'gk_diving' : 'goalkeeper_diving', 'gk_handling' : 'goalkeeper_handling', 'gk_kicking' : 'goalkeeper_kicking', 'gk_positioning' : 'goalkeeper_positioning', 'gk_reflexes' : 'goalkeeper_reflexes'})
    df = df.drop(columns=['id', 'foot', 'release_clause', 'position', 'nationality', 'club', 'team_and_contract', 'joined', 'loan_date_end', 'contract'], axis=1)
    df['best_position'] = df['best_position'].replace({'GK' : 'Goalkeeper', 'CB' : 'Center back', 'LB' : 'Left back', 'RB' : 'Right back', 'CM' : 'Centre midfield', 'CDM' : 'Centre defensive midfield', 'CAM' : 'Centre attacking midfield', 'LM' : 'Left midfield', 'RM' : 'Right midfield', 'ST' : 'Striker', 'CF' : 'Center forward', 'LW' : 'Left winger', 'RW' : 'Right winger', 'LWB' : 'Left Wing Back', 'RWB' : 'Right Wing Back'})
    df['name'] = df['name'].str.replace('.', '')
    df['height'] = df['height'].str.replace('"', '')
    df['height'] = df['height'].str.replace("'", '.')
    df['weight'] = df['weight'].str.replace('lbs', '')
    df['hits'] = df['hits'].str.replace('.', '')
    df['hits'] = df['hits'].str.replace('K', '00')
    df['hits'] = df['hits'].str.replace('M', '00000')
    columns_to_clean = ['value', 'wage']
    for column in columns_to_clean:
        df[column] = df[column].str.replace('.', '')
        df[column] = df[column].str.replace('K', '000')
        df[column] = df[column].str.replace('M', '000000')
        df[column] = df[column].str.replace('€', '')
    df['weak_foot'] = df['weak_foot'].str.replace('★', '')
    df['skill_moves'] = df['skill_moves'].str.replace('★', '')
    df['international_reputation'] = df['international_reputation'].str.replace('★', '')
    columns_to_convert = ['height', 'weight', 'value', 'wage', 'weak_foot', 'skill_moves', 'international_reputation', 'hits']
    for column in columns_to_convert:
        df[column] = pd.to_numeric(df[column], errors='coerce')
    dtypes_df = pd.DataFrame(df.dtypes, columns=['Data Types'])
    with pd.option_context('display.max_rows', None):
        print(dtypes_df)
    def convert_to_integer(value):
        try:
            return int(value)
        except (ValueError, TypeError):
            return value
    df = df.applymap(convert_to_integer)
    dtypes_df = pd.DataFrame(df.dtypes, columns=['Data Types'])
    with pd.option_context('display.max_rows', None):
        print(dtypes_df)
    return df
```
